```python
#step 1: IMPORTING REQUIRED LIBRARIES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
#step 2: IMPORTING THE DATASET
pollution=pd.read_csv("pollution.csv")
pollution.head()
```

```python
#Step 3: Identifying the features and columns
pollution.columns
#Steo 3.1: identifying datatypes
pollution.dtypes
pollution.replace({'Null':np.nan} , inplace=True)
pollution['NO2']=pollution['NO2'].astype("float")
pollution['PM10']=pollution['PM10'].astype("float")
pollution['SO2']=pollution['SO2'].astype("float")
pollution.dtypes
#fill nan
pollution['NO2'].fillna(pollution['NO2'].mean(),inplace=True)
pollution['PM10'].fillna(pollution['PM10'].mean(),inplace=True)
pollution['SO2'].fillna(pollution['SO2'].mean(),inplace=True)
#calculate total
pollution['Total_pollution']=pollution['NO2']+pollution['PM10']+pollution['SO2']
pollution
```

```python
#Step 4: EXPLORING THE CATEGORICAL VARIABLE(here,state)
states=pollution.State
print("The unique values in state category are: " + str(set(states)))
#GRAPHICAL REPRESENTATION OF STATES
sns.countplot(pollution['State'])
```

```python
#Step 5: Finding possible conclusions from the categorical data without forming any basis of the continuous data
dic={}
for s in states:
    if(s not in dic.keys()):
        dic[s]=0
    dic[s]+=1
for akey in dic:
    print('The state of {} has {} cities that are polluted.'.format(akey,dic[akey]))
smax=max(dic,key=dic.get)
smin=min(dic,key=dic.get)
print('\nthe state with highest pollution is {} while with lowest pollution is {}.'.format(smax,smin))
```
